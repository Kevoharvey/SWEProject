```python
# making a house pricing prediction model using linear regression
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
train = pd.read_csv('/Users/kevinharvey/Desktop/AI Project/train.csv')
test = pd.read_csv('/Users/kevinharvey/Desktop/AI Project/test.csv')
train.head()
test.head()

# fill numeric columns only with their respective means
num_cols = train.select_dtypes(include=[np.number]).columns
train[num_cols] = train[num_cols].fillna(train[num_cols].mean())
test[num_cols] = test[num_cols].fillna(test[num_cols].mean())

# selecting features and target variable
features = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
X_train = train[features]
y_train = train['SalePrice']
X_test = test[features]

# creating and training the model
model = LinearRegression()
model.fit(X_train, y_train)

# making predictions
predictions = model.predict(X_test)
predictions
```
