```python
# sigmoid function 
# x1= 0.8 , x2= 0.6
# w1= 0.7 , w2= 0.4
# b= -0.2
import numpy as np
z = 0.8 * 0.7 + 0.6 * 0.4 - 0.2
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

y = sigmoid(z)
error = y - 0.82
mse = 1/2 * error**2
print(mse)
```
